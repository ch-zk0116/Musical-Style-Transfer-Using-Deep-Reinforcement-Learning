### Create ADL metdata

```python
import os
import pandas as pd

# --- CONFIGURATION ---
# Set the path to the root of your ADL Piano MIDI dataset
ADL_BASE_DIR = r"C:\Users\User\Desktop\college\fyp\datasets\OriginalData\adl-piano-midi"

# Set the desired path for the output metadata CSV file
OUTPUT_CSV_PATH = r"C:\Users\User\Desktop\college\fyp\datasets\OriginalData\adl-piano-midi\adl_metadata.csv"

# --- MAIN SCRIPT ---

def create_adl_metadata():
    """
    Scans the ADL Piano MIDI dataset directory to create a metadata CSV file
    with song titles and their corresponding genre categories.
    """
    if not os.path.isdir(ADL_BASE_DIR):
        print(f"Error: The specified ADL directory does not exist: {ADL_BASE_DIR}")
        return

    metadata_list = []
    print(f"Starting scan of directory: {ADL_BASE_DIR}...")

    # Walk through the directory structure
    for root, _, files in os.walk(ADL_BASE_DIR):
        for filename in files:
            # Process only MIDI files
            if filename.lower().endswith(('.mid', '.midi')):
                try:
                    # Get the full path of the MIDI file
                    full_path = os.path.join(root, filename)

                    # --- Extract Genre Category ---
                    # Get the path relative to the base ADL directory
                    relative_path = os.path.relpath(full_path, ADL_BASE_DIR)
                    # Split the path into its components (folders/file)
                    path_components = relative_path.split(os.sep)

                    # The first component is the genre category folder
                    if len(path_components) > 1:
                        genre_category = path_components[0]
                    else:
                        # This case handles files that might be directly in the root
                        # of ADL_BASE_DIR, which is unlikely but good to handle.
                        genre_category = "unknown"
                        print(f"Warning: Could not determine genre for '{filename}' located in the root directory.")

                    # --- Extract Song Title ---
                    # Get the filename without the extension
                    song_title = os.path.splitext(filename)[0]

                    # Add the extracted data to our list
                    metadata_list.append({
                        "song_title": song_title,
                        "genre_category": genre_category
                    })
                except Exception as e:
                    print(f"An error occurred while processing file {filename}: {e}")

    if not metadata_list:
        print("No MIDI files were found or processed. The output CSV will not be created.")
        return

    # Convert the list of dictionaries to a pandas DataFrame
    df = pd.DataFrame(metadata_list)

    # Save the DataFrame to a CSV file, without the pandas index column
    try:
        # Ensure the output directory exists
        os.makedirs(os.path.dirname(OUTPUT_CSV_PATH), exist_ok=True)
        df.to_csv(OUTPUT_CSV_PATH, index=False)
        print("\n--- Success! ---")
        print(f"Metadata CSV created successfully at: {OUTPUT_CSV_PATH}")
        print(f"Total entries created: {len(df)}")
        # Print a sample of the data
        print("\nSample of the data:")
        print(df.head())
        print("\nGenre distribution:")
        print(df['genre_category'].value_counts())

    except Exception as e:
        print(f"\nError: Could not save the CSV file. Reason: {e}")

if __name__ == "__main__":
    create_adl_metadata()
```

# Data Preprocessing

### Validation Preprocessing

```python
import pretty_midi
import mido
import os
import pandas as pd
import json
import shutil

class MidiProcessor:
    """
    Parses, validates, and processes a MIDI file.

    This class loads a MIDI file, performs several validation checks,
    and filters it to contain only acoustic piano instruments with valid notes.
    The output is a new, cleaned pretty_midi object.
    """
    def __init__(self, filepath, verbose=False):
        """
        Initializes the processor and runs the validation and filtering.

        Args:
            filepath (str): The path to the MIDI file.
            verbose (bool, optional): If True, prints validation errors,
                                      warnings, and other info to the console.
                                      Defaults to False.
        """
        self.filepath = filepath
        self.verbose = verbose  # <<< MODIFIED: Store the verbose flag
        self.pm_obj = None
        self.mido_obj = None
        self.is_valid_midi = True
        self.error_messages = []
        self.processed_pm_obj = None

        self._parse_and_validate()

    def _add_error(self, message, is_critical=True):
        """Internal method to log an error and update the validation status."""
        self.error_messages.append(message)
        if is_critical:
            self.is_valid_midi = False
        
        # <<< MODIFIED: Only print if verbose is enabled
        if self.verbose:
            print(f"Validation {'Error' if is_critical else 'Warning'} for {self.filepath}: {message}")

    def _parse_and_validate(self):
        if not os.path.exists(self.filepath):
            self._add_error("File does not exist.")
            return

        try:
            self.pm_obj = pretty_midi.PrettyMIDI(self.filepath)
        except Exception as e:
            self._add_error(f"Not parsable by pretty_midi: {e}")
            return

        try:
            # clip=True can help with some malformed files
            self.mido_obj = mido.MidiFile(self.filepath, clip=True)
            if self.mido_obj.type not in [0, 1, 2]:
                self._add_error(f"Invalid MIDI file type: {self.mido_obj.type}")
            if self.mido_obj.ticks_per_beat <= 0:
                self._add_error(f"Invalid MIDI resolution (ticks_per_beat): {self.mido_obj.ticks_per_beat}")
        except Exception as e:
            # Non-critical if pretty_midi worked
            self._add_error(f"Error during mido parsing or SMF checks: {e}", is_critical=False)
            self.mido_obj = None

        # if pretty_midi failed or a critical mido check failed
        if not self.is_valid_midi:
            return

        if self.mido_obj:
            all_tracks_have_eot = True
            # For Type 0/1/2, iterate through tracks.
            tracks_to_check = self.mido_obj.tracks if self.mido_obj.type in [1, 2] else [self.mido_obj]

            for track in tracks_to_check:
                messages = list(track)
                if messages and not any(msg.type == 'end_of_track' for msg in messages):
                    all_tracks_have_eot = False
                    break
            if not all_tracks_have_eot:
                 # Non-critical
                self._add_error("One or more tracks may be missing end_of_track events.", is_critical=False)

        # Filter for Piano Only & Check for Notes
        piano_instruments_only = []
        has_any_piano = False
        if not self.pm_obj.instruments:
            self._add_error("No instruments found in the MIDI file (pretty_midi).")
            return

        for instrument in self.pm_obj.instruments:
            # Only keep Acoustic Grand Piano (0) and Bright Acoustic Piano (1)
            if not instrument.is_drum and (0 <= instrument.program <= 1):
                valid_notes = [note for note in instrument.notes if note.end > note.start]
                
                # <<< MODIFIED: Only print if verbose is enabled
                if self.verbose and len(valid_notes) < len(instrument.notes):
                    print(f"Info for {self.filepath}: Instrument '{instrument.name}' (Program {instrument.program}) "
                          f"had {len(instrument.notes) - len(valid_notes)} zero-duration notes removed.")
                
                instrument.notes = valid_notes

                if instrument.notes:
                    piano_instruments_only.append(instrument)
                    has_any_piano = True

        if not has_any_piano:
            self._add_error("No acoustic piano instruments with notes found after filtering.")
            return

        # Create a new PrettyMIDI object for the processed output
        # Get initial tempo carefully
        initial_tempo = 120.0  # Default
        try:
            tempo_changes_time, tempo_changes_bpm = self.pm_obj.get_tempo_changes()
            if tempo_changes_bpm.size > 0:
                initial_tempo = tempo_changes_bpm[0]
        except ValueError:
             self._add_error("Could not determine tempo from MIDI, using default 120.", is_critical=False)


        self.processed_pm_obj = pretty_midi.PrettyMIDI(resolution=self.pm_obj.resolution,
                                                       initial_tempo=initial_tempo)
        self.processed_pm_obj.instruments = piano_instruments_only

        total_piano_notes = sum(len(instr.notes) for instr in self.processed_pm_obj.instruments)
        if total_piano_notes == 0:
            self._add_error("Filtered piano instruments collectively have no notes.")
            self.processed_pm_obj = None
            return

        # <<< MODIFIED: Only print if verbose is enabled
        if self.is_valid_midi and self.verbose:
             print(f"File {self.filepath} passed initial validation and piano filtering.")

    def get_processed_midi(self):
        """Returns the cleaned and filtered pretty_midi object, or None if validation failed."""
        if self.is_valid_midi and self.processed_pm_obj:
            return self.processed_pm_obj
        return None

    def get_errors(self):
        """Returns a list of all error and warning messages collected during processing."""
        return self.error_messages

# --- Configuration ---
BASE_INPUT_DIR = r"C:\Users\User\Desktop\college\fyp\datasets\OriginalData"
BASE_OUTPUT_DIR = r"C:\Users\User\Desktop\college\fyp\datasets\PreprocessedData\validFiles"
BATCH_PRINT_INTERVAL = 1000
ADL_DIR_NAME, PIAST_DIR_NAME, PIJAMA_DIR_NAME = "adl-piano-midi", "piast", "piJAMA"
PIAST_AT_CSV_PATH = os.path.join(BASE_INPUT_DIR, PIAST_DIR_NAME, "Hayeonbang_PIAST_dataset", "piast_at", "at_text.csv")
PIAST_YT_JSON_PATH = os.path.join(BASE_INPUT_DIR, PIAST_DIR_NAME, "Hayeonbang_PIAST_dataset", "piast_yt", "yt_text.json")

# --- High-Performance Metadata Loading ---
def load_piast_metadata():
    """
    Loads PIAST metadata into both a DataFrame for flexible searching and
    a dictionary (hash map) for O(1) exact-match lookups.
    Returns a dictionary for each dataset containing the df, map, and id column name.
    """
    piast_at_meta, piast_yt_meta = None, None
    
    try:
        df_at = pd.read_csv(PIAST_AT_CSV_PATH)
        if 'AudioFile' in df_at.columns:
            df_at['AudioFile'] = df_at['AudioFile'].astype(str)
            map_at = pd.Series(df_at.to_dict('records'), index=df_at.AudioFile).to_dict()
            piast_at_meta = {'df': df_at, 'map': map_at, 'id_col': 'AudioFile'}
            print(f"Successfully loaded and indexed PIAST-AT metadata.")
    except Exception as e:
        print(f"Warning: Could not load or index PIAST-AT metadata: {e}")

    try:
        df_yt = pd.read_json(PIAST_YT_JSON_PATH)
        if 'track_id' in df_yt.columns:
            df_yt['track_id'] = df_yt['track_id'].astype(str)
            map_yt = pd.Series(df_yt.to_dict('records'), index=df_yt.track_id).to_dict()
            piast_yt_meta = {'df': df_yt, 'map': map_yt, 'id_col': 'track_id'}
            print(f"Successfully loaded and indexed PIAST-YT metadata.")
    except Exception as e:
        print(f"Warning: Could not load or index PIAST-YT metadata: {e}")
        
    return piast_at_meta, piast_yt_meta

# --- Centralized Metadata Lookup Function ---
def find_piast_metadata(midi_title, metadata_bundle):
    """
    Finds metadata for a MIDI file using a fast hash map and an optimized search fallback.
    """
    if not metadata_bundle: return None
    metadata_map = metadata_bundle['map']
    if midi_title in metadata_map:
        return metadata_map[midi_title]
    if "_part_" not in midi_title:
        return None
    try:
        midi_base, midi_part = midi_title.split("_part_", 1)
        metadata_df = metadata_bundle['df']
        id_col = metadata_bundle['id_col']
        candidates = metadata_df[
            metadata_df[id_col].str.startswith(midi_base, na=False) &
            metadata_df[id_col].str.endswith(f"_part_{midi_part}", na=False)
        ]
        if not candidates.empty:
            return candidates.iloc[0].to_dict()
    except (ValueError, IndexError):
        return None
    return None

# --- Jazz Classification Helpers ---
def is_adl_jazz(filepath, adl_base_path):
    try:
        relative_path = os.path.relpath(filepath, adl_base_path)
        return relative_path.split(os.sep)[0].lower() == "jazz"
    except (ValueError, IndexError):
        return False

def is_piast_at_jazz(midi_title, piast_at_meta):
    record = find_piast_metadata(midi_title, piast_at_meta)
    if record:
        for col in ["Agreements_3", "Agreements_2"]:
            if col in record and pd.notna(record[col]) and "jazz" in str(record[col]).lower():
                return True
    return False

def is_piast_yt_jazz(midi_title, piast_yt_meta):
    record = find_piast_metadata(midi_title, piast_yt_meta)
    if record:
        tags = record.get("tag", [])
        if tags and isinstance(tags, list) and len(tags) > 0:
            return "jazz" in str(tags[0]).lower()
    return False

# --- Main Processing Loop ---
def process_datasets():
    # MODIFIED: Initialize new counters
    total_files_processed = 0
    total_valid_files_saved = 0
    
    piast_at_meta, piast_yt_meta = load_piast_metadata()

    datasets = [
        {"name": "adl", "path": os.path.join(BASE_INPUT_DIR, ADL_DIR_NAME), "type": "adl"},
        {"name": "piast", "path": os.path.join(BASE_INPUT_DIR, PIAST_DIR_NAME, "Hayeonbang_PIAST_dataset"), "type": "piast"},
        {"name": "piJAMA", "path": os.path.join(BASE_INPUT_DIR, PIJAMA_DIR_NAME, "midi_kong"), "type": "pijama"},
    ]
    for d_info in datasets:
        print(f"\n--- Processing Dataset: {d_info['name']} ---")
        if not os.path.exists(d_info['path']):
            print(f"Path not found: {d_info['path']}. Skipping."); continue
        
        if d_info['type'] == 'piast':
            for piast_type, sub_path in {"at": "piast_at", "yt": "piast_yt"}.items():
                midi_dir = os.path.join(d_info['path'], sub_path, "midi")
                print(f"  -- Processing PIAST sub-type: {sub_path} --")
                if not os.path.exists(midi_dir): continue
                for filename in os.listdir(midi_dir):
                    if not filename.lower().endswith((".mid", ".midi")): continue
                    
                    # MODIFIED: Increment processed count for every file attempted
                    total_files_processed += 1
                    
                    filepath, title = os.path.join(midi_dir, filename), os.path.splitext(filename)[0]
                    processor = MidiProcessor(filepath)
                    if pm_object := processor.get_processed_midi():
                        is_jazz = is_piast_at_jazz(title, piast_at_meta) if piast_type == "at" else is_piast_yt_jazz(title, piast_yt_meta)
                        genre_folder = "jazz" if is_jazz else "non_jazz"
                        output_dir = os.path.join(BASE_OUTPUT_DIR, d_info['name'], sub_path, genre_folder)
                        os.makedirs(output_dir, exist_ok=True)
                        pm_object.write(os.path.join(output_dir, filename))
                        
                        # MODIFIED: Increment saved count only for valid files
                        total_valid_files_saved += 1
                    
                    # MODIFIED: Report status based on total processed files
                    if total_files_processed > 0 and total_files_processed % BATCH_PRINT_INTERVAL == 0:
                        total_rejected = total_files_processed - total_valid_files_saved
                        print(f"-> Status: {total_files_processed} files processed | "
                              f"{total_valid_files_saved} valid | {total_rejected} rejected.")
            continue
            
        # Handling for ADL and piJAMA
        for root, _, files in os.walk(d_info['path']):
            for filename in files:
                if not filename.lower().endswith((".mid", ".midi")): continue
                
                # MODIFIED: Increment processed count
                total_files_processed += 1
                
                filepath = os.path.join(root, filename)
                processor = MidiProcessor(filepath)
                if pm_object := processor.get_processed_midi():
                    if d_info['type'] == 'pijama':
                        output_dir = os.path.join(BASE_OUTPUT_DIR, d_info['name'])
                    else: # ADL
                        genre_folder = "jazz" if is_adl_jazz(filepath, d_info['path']) else "non_jazz"
                        output_dir = os.path.join(BASE_OUTPUT_DIR, d_info['name'], genre_folder)
                    os.makedirs(output_dir, exist_ok=True)
                    pm_object.write(os.path.join(output_dir, filename))
                    
                    # MODIFIED: Increment saved count
                    total_valid_files_saved += 1
                
                # MODIFIED: Report status based on total processed files
                if total_files_processed > 0 and total_files_processed % BATCH_PRINT_INTERVAL == 0:
                    total_rejected = total_files_processed - total_valid_files_saved
                    print(f"-> Status: {total_files_processed} files processed | "
                          f"{total_valid_files_saved} valid | {total_rejected} rejected.")

    # MODIFIED: Return final counts for the summary
    return total_files_processed, total_valid_files_saved


os.makedirs(BASE_OUTPUT_DIR, exist_ok=True)
# MODIFIED: Capture the final counts from the function
processed_count, saved_count = process_datasets()
rejected_count = processed_count - saved_count

# MODIFIED: Print a detailed final summary
print("\n--- All dataset processing complete. ---")
print("\n--- Final Summary ---")
print(f"Total files processed (attempted): {processed_count}")
print(f"Total valid files saved: {saved_count}")
print(f"Total files rejected (invalid/filtered): {rejected_count}")
```

```python
# metadata creation script
import pretty_midi
import os
import pandas as pd
import json
import numpy as np

# ==============================================================================
# SCRIPT CONFIGURATION
# ==============================================================================
BASE_INPUT_DIR = r"C:\Users\User\Desktop\college\fyp\datasets\OriginalData"
BASE_OUTPUT_DIR = r"C:\Users\User\Desktop\college\fyp\datasets\PreprocessedData\validFiles"
METADATA_OUTPUT_DIR = r"C:\Users\User\Desktop\college\fyp\datasets\PreprocessedData\metadata"
os.makedirs(METADATA_OUTPUT_DIR, exist_ok=True)

# Dataset & Metadata Config
ADL_DIR_NAME, PIAST_DIR_NAME, PIJAMA_DIR_NAME = "adl-piano-midi", "piast", "piJAMA"
ADL_ORIGINAL_METADATA_CSV_PATH = os.path.join(BASE_INPUT_DIR, ADL_DIR_NAME, "adl_metadata.csv")
PIAST_AT_CSV_PATH = os.path.join(BASE_INPUT_DIR, PIAST_DIR_NAME, "Hayeonbang_PIAST_dataset", "piast_at", "at_text.csv")
PIAST_YT_JSON_PATH = os.path.join(BASE_INPUT_DIR, PIAST_DIR_NAME, "Hayeonbang_PIAST_dataset", "piast_yt", "yt_text.json")

# Metadata Column Definitions
# <<< MODIFIED: Removed "Key Signature" and "Time Signature" columns. >>>
common_cols = [
    "Length (s)", "Note Count", "Avg Notes/Min", "BPM", "Tempo Changes Count", 
    "Min Velocity", "Max Velocity", "Unique Pitches Count"
]
adl_cols = ["Title", "Genre"] + common_cols
piast_yt_cols = ["Title", "Tags"] + common_cols
piast_at_cols = ["Title", "Agreements_2", "Agreements_3"] + common_cols

# ==============================================================================
# HELPER FUNCTIONS (Identical to other script for consistency)
# ==============================================================================
def load_piast_metadata():
    """Loads PIAST metadata into DataFrames and hash maps for fast lookups."""
    piast_at_meta, piast_yt_meta = None, None
    try:
        df_at = pd.read_csv(PIAST_AT_CSV_PATH)
        if 'AudioFile' in df_at.columns:
            df_at['AudioFile'] = df_at['AudioFile'].astype(str)
            map_at = pd.Series(df_at.to_dict('records'), index=df_at.AudioFile).to_dict()
            piast_at_meta = {'df': df_at, 'map': map_at, 'id_col': 'AudioFile'}
            print(f"Successfully loaded and indexed PIAST-AT metadata.")
    except Exception as e: print(f"Warning: Could not load or index PIAST-AT metadata: {e}")
    try:
        df_yt = pd.read_json(PIAST_YT_JSON_PATH)
        if 'track_id' in df_yt.columns:
            df_yt['track_id'] = df_yt['track_id'].astype(str)
            map_yt = pd.Series(df_yt.to_dict('records'), index=df_yt.track_id).to_dict()
            piast_yt_meta = {'df': df_yt, 'map': map_yt, 'id_col': 'track_id'}
            print(f"Successfully loaded and indexed PIAST-YT metadata.")
    except Exception as e: print(f"Warning: Could not load or index PIAST-YT metadata: {e}")
    return piast_at_meta, piast_yt_meta

def find_piast_metadata(midi_title, metadata_bundle):
    """Finds metadata using a fast hash map and an optimized search fallback."""
    if not metadata_bundle: return None
    if midi_title in metadata_bundle['map']: return metadata_bundle['map'][midi_title]
    if "_part_" not in midi_title: return None
    try:
        midi_base, midi_part = midi_title.split("_part_", 1)
        candidates = metadata_bundle['df'][
            metadata_bundle['df'][metadata_bundle['id_col']].str.startswith(midi_base, na=False) &
            metadata_bundle['df'][metadata_bundle['id_col']].str.endswith(f"_part_{midi_part}", na=False)
        ]
        if not candidates.empty: return candidates.iloc[0].to_dict()
    except (ValueError, IndexError): return None
    return None

def load_adl_original_metadata(adl_csv_path):
    if os.path.exists(adl_csv_path):
        try:
            df = pd.read_csv(adl_csv_path)
            if {'song_title', 'genre_category'}.issubset(df.columns): return df
        except Exception as e: print(f"Error loading ADL metadata: {e}")
    return None

# <<< MODIFIED: Removed all logic for key and time signatures. >>>
def extract_midi_features(midi_filepath):
    """
    Extracts features from a MIDI file for analysis.
    """
    # Initialize features with null/zero values
    features = {k: np.nan for k in ["BPM", "Min Velocity", "Max Velocity"]}
    features.update({k: 0 for k in [
        "Length (s)", "Note Count", "Avg Notes/Min", "Tempo Changes Count", 
        "Unique Pitches Count"
    ]})

    try:
        pm = pretty_midi.PrettyMIDI(midi_filepath)
        
        # Note Count and Rate Calculations
        length_sec = pm.get_end_time()
        features["Length (s)"] = length_sec
        
        notes = [n for i in pm.instruments for n in i.notes]
        note_count = len(notes)
        features["Note Count"] = note_count

        if length_sec > 0:
            features["Avg Notes/Min"] = note_count / (length_sec / 60.0)

        if notes:
            velocities = [n.velocity for n in notes]
            features["Min Velocity"], features["Max Velocity"] = min(velocities), max(velocities)
            features["Unique Pitches Count"] = len(set(n.pitch for n in notes))

        # BPM Extraction Logic
        bpm_value = np.nan
        try:
            if notes: bpm_value = pm.estimate_tempo()
        except Exception: bpm_value = np.nan

        tempos_list = pm.get_tempo_changes()[1]
        features["Tempo Changes Count"] = len(tempos_list)

        if np.isnan(bpm_value) and len(tempos_list) > 0:
            bpm_value = tempos_list[0]
        
        features["BPM"] = bpm_value

    except Exception as e:
        print(f"ERROR: Could not process features for {midi_filepath}: {e}")
    
    return features
    
# ==============================================================================
# MAIN METADATA CREATION LOGIC
# ==============================================================================
def create_metadata_files():
    print("\n--- Starting Metadata File Creation ---")
    adl_orig_df = load_adl_original_metadata(ADL_ORIGINAL_METADATA_CSV_PATH)
    piast_at_meta, piast_yt_meta = load_piast_metadata()

    targets = [
        {"id": "pijama_jazz", "path": os.path.join(BASE_OUTPUT_DIR, PIJAMA_DIR_NAME), "csv": "pijama_jazz_metadata.csv", "cols": adl_cols, "type": "pijama"},
        {"id": "adl_jazz", "path": os.path.join(BASE_OUTPUT_DIR, "adl", "jazz"), "csv": "adl_jazz_metadata.csv", "cols": adl_cols, "type": "adl", "genre": "Jazz"},
        {"id": "adl_non_jazz", "path": os.path.join(BASE_OUTPUT_DIR, "adl", "non_jazz"), "csv": "adl_non_jazz_metadata.csv", "cols": adl_cols, "type": "adl", "genre": "Non-Jazz"},
        {"id": "piast_at_jazz", "path": os.path.join(BASE_OUTPUT_DIR, "piast", "piast_at", "jazz"), "csv": "piast_at_jazz_metadata.csv", "cols": piast_at_cols, "type": "piast_at"},
        {"id": "piast_at_non_jazz", "path": os.path.join(BASE_OUTPUT_DIR, "piast", "piast_at", "non_jazz"), "csv": "piast_at_non_jazz_metadata.csv", "cols": piast_at_cols, "type": "piast_at"},
        {"id": "piast_yt_jazz", "path": os.path.join(BASE_OUTPUT_DIR, "piast", "piast_yt", "jazz"), "csv": "piast_yt_jazz_metadata.csv", "cols": piast_yt_cols, "type": "piast_yt"},
        {"id": "piast_yt_non_jazz", "path": os.path.join(BASE_OUTPUT_DIR, "piast", "piast_yt", "non_jazz"), "csv": "piast_yt_non_jazz_metadata.csv", "cols": piast_yt_cols, "type": "piast_yt"},
    ]
    for t in targets:
        print(f"\nProcessing target: {t['id']}")
        if not os.path.exists(t['path']): print(f"Warning: Path not found, skipping: {t['path']}"); continue
        
        all_rows = []
        for filename in os.listdir(t['path']):
            if not filename.lower().endswith((".mid", ".midi")): continue
            filepath, title = os.path.join(t['path'], filename), os.path.splitext(filename)[0]
            row = {"Title": title, **extract_midi_features(filepath)}

            if t['type'] == 'pijama':
                row['Genre'] = 'Jazz'
            elif t['type'] == 'adl':
                genre = t.get('genre', 'N/A')
                if adl_orig_df is not None:
                    match = adl_orig_df[adl_orig_df['song_title'].astype(str).str.lower() == title.lower()]
                    if not match.empty: genre = match.iloc[0].get('genre_category', genre)
                row['Genre'] = genre
            elif t['type'] == 'piast_at':
                record = find_piast_metadata(title, piast_at_meta)
                row['Agreements_2'] = record.get('Agreements_2', 'N/A') if record else 'N/A'
                row['Agreements_3'] = record.get('Agreements_3', 'N/A') if record else 'N/A'
            elif t['type'] == 'piast_yt':
                record = find_piast_metadata(title, piast_yt_meta)
                tags = record.get('tag', []) if record else []
                row['Tags'] = tags[0] if tags and isinstance(tags, list) and len(tags) > 0 else 'N/A'
            
            all_rows.append(row)

        if all_rows:
            df = pd.DataFrame(all_rows, columns=t['cols'])
            df.to_csv(os.path.join(METADATA_OUTPUT_DIR, t['csv']), index=False, encoding='utf-8')
            print(f"Successfully created: {t['csv']} with {len(df)} entries.")
            
    print("\n--- Metadata file creation complete. ---")

if __name__ == "__main__":
    create_metadata_files()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np
import json
from IPython.display import display
from scipy import stats # For statistical tests

class DatasetAnalyzer:
    def __init__(self, metadata_dir, piast_at_tag_list_path):
            print(f"\n--- Initializing DatasetAnalyzer for: {metadata_dir} ---")
            self.metadata_dir = metadata_dir
            self.piast_at_tag_list_path = piast_at_tag_list_path

            # Define these attributes BEFORE calling _load_all_metadata
            self.numerical_cols_for_analysis = [
                "Length (s)",
                "BPM",
                "Note Count",
                "Avg Notes/Min",
                "Min Velocity", "Max Velocity",
                "Unique Pitches Count"
            ]
            self.numerical_cols_to_plot = [
                "Length (s)",
                "BPM",
                "Note Count",
                "Avg Notes/Min",
                "Min Velocity", "Max Velocity",
                "Unique Pitches Count"
            ]
            
            # Now call _load_all_metadata
            self.combined_metadata_df, self.piast_at_defined_tags = self._load_all_metadata()


    def _load_all_metadata(self):
        all_dfs = []
        print(f"Looking for metadata files in: {self.metadata_dir}")

        files_to_load_info = {
            "adl_jazz_metadata.csv": {"broad_genre": "Jazz", "dataset": "ADL"},
            "adl_non_jazz_metadata.csv": {"broad_genre": "Non-Jazz", "dataset": "ADL"},
            "piast_at_jazz_metadata.csv": {"broad_genre": "Jazz", "dataset": "PiAST-AT"},
            "piast_at_non_jazz_metadata.csv": {"broad_genre": "Non-Jazz", "dataset": "PiAST-AT"},
            "piast_yt_jazz_metadata.csv": {"broad_genre": "Jazz", "dataset": "PiAST-YT"},
            "piast_yt_non_jazz_metadata.csv": {"broad_genre": "Non-Jazz", "dataset": "PiAST-YT"},
            "pijama_jazz_metadata.csv": {"broad_genre": "Jazz", "dataset": "PiJAMA"},
        }

        for filename, info in files_to_load_info.items():
            filepath = os.path.join(self.metadata_dir, filename)
            if os.path.exists(filepath):
                try:
                    df = pd.read_csv(filepath)
                    df['Broad_Genre'] = info['broad_genre']
                    df['Dataset_Source'] = info['dataset']
                    all_dfs.append(df)
                    print(f"Loaded: {filename} with {len(df)} rows.")
                except Exception as e:
                    print(f"Error loading {filename}: {e}")
            else:
                print(f"Warning: File not found {filepath}")

        if not all_dfs:
            print("No metadata files loaded for this analyzer instance.")
            return None, None

        combined_df = pd.concat(all_dfs, ignore_index=True)
        print(f"Total combined rows for this instance: {len(combined_df)}")

        # Ensure all numerical columns for analysis are actually numeric, coercing errors
        for col in self.numerical_cols_for_analysis + ['Length (s)']: # Include Length for overall df
             if col in combined_df.columns:
                combined_df[col] = pd.to_numeric(combined_df[col], errors='coerce')
             else:
                print(f"Warning: Column '{col}' expected for numerical analysis not found in combined data.")


        piast_at_tag_list = None
        if os.path.exists(self.piast_at_tag_list_path):
            try:
                with open(self.piast_at_tag_list_path, 'r') as f:
                    piast_at_tag_list = json.load(f)
                print(f"Successfully loaded PiAST-AT tag list with {len(piast_at_tag_list)} tags.")
            except Exception as e:
                print(f"Error loading PiAST-AT tag list: {e}")
        else:
            print(f"Warning: PiAST-AT tag list not found at {self.piast_at_tag_list_path}")

        return combined_df, piast_at_tag_list

    def _get_outlier_info(self, series):
        if series.empty or series.isnull().all():
            return 0, np.nan, np.nan, np.nan, np.nan
        Q1 = series.quantile(0.25)
        Q3 = series.quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = series[(series < lower_bound) | (series > upper_bound)]
        non_outlier_series = series[(series >= lower_bound) & (series <= upper_bound)]
        non_outliers_min = non_outlier_series.min() if not non_outlier_series.empty else np.nan
        non_outliers_max = non_outlier_series.max() if not non_outlier_series.empty else np.nan
        return len(outliers), lower_bound, upper_bound, non_outliers_min, non_outliers_max

    # --- Plotting Functions (plot_adl_genre_distribution, etc. - keep them as they are) ---
    def plot_adl_genre_distribution(self):
        if self.combined_metadata_df is None or self.combined_metadata_df.empty:
            print("No data loaded to plot ADL genre distribution.")
            return
        print("\nPlotting ADL Genre Distribution...")
        adl_df_plot = self.combined_metadata_df[self.combined_metadata_df['Dataset_Source'] == 'ADL'].copy()
        if not adl_df_plot.empty and 'Genre' in adl_df_plot.columns:
            plt.figure(figsize=(12, 7))
            adl_genre_counts = adl_df_plot['Genre'].value_counts()
            ax = sns.barplot(x=adl_genre_counts.index, y=adl_genre_counts.values)
            plt.title(f'ADL Dataset: Specific Genre Distribution\n(Source: {self.metadata_dir})')
            plt.xlabel('Specific Genre'); plt.ylabel('Count')
            plt.xticks(rotation=45, ha='right')
            for i, v in enumerate(adl_genre_counts.values):
                ax.text(i, v + 0.5, str(v), color='black', ha='center', va='bottom', fontsize=9)
            plt.tight_layout(); plt.show()
        else: print("ADL data or 'Genre' column not found for plotting.")

    def plot_piast_yt_tags_distribution(self):
        if self.combined_metadata_df is None or self.combined_metadata_df.empty:
            print("No data loaded to plot PIAST-YT tags distribution.")
            return
        print("\nPlotting PiAST-YT Tags Distribution...")
        piast_yt_df_plot = self.combined_metadata_df[self.combined_metadata_df['Dataset_Source'] == 'PiAST-YT'].copy()
        if not piast_yt_df_plot.empty and 'Tags' in piast_yt_df_plot.columns:
            def split_piast_yt_tags(tag_entry):
                if isinstance(tag_entry, list) and len(tag_entry) > 0 and isinstance(tag_entry[0], str):
                    return [tag.strip().lower() for tag in tag_entry[0].split(',')]
                elif isinstance(tag_entry, str): return [tag.strip().lower() for tag in tag_entry.split(',')]
                return []
            piast_yt_df_plot.loc[:, 'Individual_Tags'] = piast_yt_df_plot['Tags'].apply(split_piast_yt_tags)
            exploded_tags_yt = piast_yt_df_plot.explode('Individual_Tags')
            exploded_tags_yt = exploded_tags_yt[exploded_tags_yt['Individual_Tags'] != '']
            for broad_genre_type in ['Jazz', 'Non-Jazz']:
                subset_df_yt = exploded_tags_yt[exploded_tags_yt['Broad_Genre'] == broad_genre_type]
                if not subset_df_yt.empty and 'Individual_Tags' in subset_df_yt.columns:
                    tag_counts_yt = subset_df_yt['Individual_Tags'].value_counts().nlargest(20)
                    if not tag_counts_yt.empty:
                        plt.figure(figsize=(14, 8))
                        ax = sns.barplot(x=tag_counts_yt.index, y=tag_counts_yt.values)
                        plt.title(f'PiAST-YT ({broad_genre_type}): Top 20 Individual Tags\n(Source: {self.metadata_dir})')
                        plt.xlabel('Individual Tag'); plt.ylabel('Count')
                        plt.xticks(rotation=60, ha='right')
                        for i, v in enumerate(tag_counts_yt.values):
                            ax.text(i, v + 0.2, str(v), color='black', ha='center', va='bottom', fontsize=9)
                        plt.tight_layout(); plt.show()
                    else: print(f"No individual tags to plot for PiAST-YT {broad_genre_type}")
        else: print("PiAST-YT data or 'Tags' column not found for plotting.")

    def plot_piast_at_agreement_distribution(self):
        if self.combined_metadata_df is None or self.combined_metadata_df.empty or self.piast_at_defined_tags is None:
            print("No data or PiAST-AT defined tags loaded to plot PIAST-AT agreement distribution.")
            return
        print("\nPlotting PiAST-AT Agreement Distribution...")
        piast_at_df_plot = self.combined_metadata_df[self.combined_metadata_df['Dataset_Source'] == 'PiAST-AT'].copy()
        if not piast_at_df_plot.empty and ('Agreements_2' in piast_at_df_plot.columns or 'Agreements_3' in piast_at_df_plot.columns):
            all_valid_agreement_tags = []
            for index, row in piast_at_df_plot.iterrows():
                broad_genre = row['Broad_Genre']
                tags_from_row_for_df = []
                for col_agree in ['Agreements_2', 'Agreements_3']:
                    if col_agree in row and pd.notna(row[col_agree]):
                        agreements_str = str(row[col_agree])
                        potential_tags = [tag.strip() for tag in agreements_str.split(',')]
                        for p_tag in potential_tags:
                            if p_tag in self.piast_at_defined_tags:
                                tags_from_row_for_df.append({'Agreement_Tag': p_tag, 'Broad_Genre': broad_genre})
                all_valid_agreement_tags.extend(tags_from_row_for_df)
            if all_valid_agreement_tags:
                df_agreement_tags = pd.DataFrame(all_valid_agreement_tags)
                for broad_genre_type in ['Jazz', 'Non-Jazz']:
                    subset_df_at = df_agreement_tags[df_agreement_tags['Broad_Genre'] == broad_genre_type]
                    if not subset_df_at.empty and 'Agreement_Tag' in subset_df_at.columns:
                        agreement_counts_at = subset_df_at['Agreement_Tag'].value_counts()
                        if not agreement_counts_at.empty:
                            plt.figure(figsize=(14, 8))
                            ax = sns.barplot(x=agreement_counts_at.index, y=agreement_counts_at.values)
                            plt.title(f'PiAST-AT ({broad_genre_type}): Valid Agreement Tags\n(Source: {self.metadata_dir})')
                            plt.xlabel('Agreement Tag (from defined list)'); plt.ylabel('Count')
                            plt.xticks(rotation=60, ha='right')
                            for i, v in enumerate(agreement_counts_at.values):
                                ax.text(i, v + 0.2, str(v), color='black', ha='center', va='bottom', fontsize=9)
                            plt.tight_layout(); plt.show()
                        else: print(f"No valid agreement tags to plot for PiAST-AT {broad_genre_type}")
            else: print("No valid agreement tags found in PiAST-AT data to plot.")
        elif not self.piast_at_defined_tags: print("PiAST-AT defined tag list not loaded. Skipping PiAST-AT agreement plots.")
        else: print("PiAST-AT data or Agreement columns not found for plotting.")

    def _plot_single_numerical_feature(self, data_df, column_name, plot_title_prefix="Feature"):
        plot_df_col = data_df.copy()
        if plot_df_col.empty or column_name not in plot_df_col.columns:
            print(f"Not enough data or column '{column_name}' missing to plot for {plot_title_prefix}.")
            return
        plot_df_col[column_name] = pd.to_numeric(plot_df_col[column_name], errors='coerce')
        plot_df_col.dropna(subset=[column_name], inplace=True)
        if plot_df_col.empty:
            print(f"No valid data for column '{column_name}' in {plot_title_prefix} after dropping NaNs.")
            return

        plt.figure(figsize=(16, 7))
        ax_box = plt.subplot(1, 2, 1)
        sns.boxplot(x='Broad_Genre', y=column_name, data=plot_df_col, order=['Jazz', 'Non-Jazz'], ax=ax_box)
        outlier_texts = []
        for genre in ['Jazz', 'Non-Jazz']:
            genre_series = plot_df_col[plot_df_col['Broad_Genre'] == genre][column_name]
            if not genre_series.empty:
                count, low_b, upp_b, non_out_min, non_out_max = self._get_outlier_info(genre_series)
                outlier_texts.append(f"{genre}: {count} outliers. Non-outlier: [{non_out_min:.2f}-{non_out_max:.2f}] (B: [{low_b:.2f},{upp_b:.2f}])")
        ax_box.set_title(f'{plot_title_prefix}: Box Plot of {column_name}\n(Jazz vs. Non-Jazz)\n' + '\n'.join(outlier_texts), fontsize=10)
        ax_box.set_xlabel('Broad Genre'); ax_box.set_ylabel(column_name)

        ax_hist = plt.subplot(1, 2, 2)
        sns.histplot(data=plot_df_col, x=column_name, hue='Broad_Genre', kde=True, hue_order=['Jazz', 'Non-Jazz'], element="step", stat="density", common_norm=False, ax=ax_hist)
        skew_texts = []
        for genre in ['Jazz', 'Non-Jazz']:
            genre_series = plot_df_col[plot_df_col['Broad_Genre'] == genre][column_name]
            if not genre_series.empty: skew_texts.append(f"{genre} Skew: {genre_series.skew():.2f}")
        ax_hist.set_title(f'{plot_title_prefix}: Hist & KDE of {column_name}\n(Jazz vs. Non-Jazz)\n' + ', '.join(skew_texts), fontsize=10)
        ax_hist.set_xlabel(column_name); ax_hist.set_ylabel('Density')
        plt.tight_layout(); plt.show()

    def plot_numerical_feature_distributions(self):
        if self.combined_metadata_df is None or self.combined_metadata_df.empty:
            print("No data loaded to plot numerical feature distributions.")
            return
        print("\nPlotting Numerical Feature Distributions...")
        if 'Broad_Genre' not in self.combined_metadata_df.columns:
            print("Error: 'Broad_Genre' column missing. Cannot create Jazz vs Non-Jazz plots.")
            return

        piast_at_length_df = self.combined_metadata_df[self.combined_metadata_df['Dataset_Source'] == 'PiAST-AT']
        print("\nPlotting PiAST-AT Length (s)...")
        self._plot_single_numerical_feature(piast_at_length_df, "Length (s)", "PiAST-AT")

        general_length_df = self.combined_metadata_df[self.combined_metadata_df['Dataset_Source'] != 'PiAST-AT']
        print("\nPlotting General (Excl. PiAST-AT) Length (s)...")
        self._plot_single_numerical_feature(general_length_df, "Length (s)", "General (Excl. PiAST-AT)")

        for col_num in self.numerical_cols_to_plot: # Use self.numerical_cols_to_plot
            if col_num == "Length (s)": continue
            print(f"\nPlotting Overall {col_num}...")
            self._plot_single_numerical_feature(self.combined_metadata_df.copy(), col_num, "Overall")

    # <<< NEW METHODS FOR CORRELATION AND STATISTICAL SIGNIFICANCE >>>
    def perform_correlation_analysis(self):
        if self.combined_metadata_df is None or self.combined_metadata_df.empty:
            print("No data loaded to perform correlation analysis.")
            return

        print(f"\n--- Performing Correlation Analysis on Numerical Features for data from: {self.metadata_dir} ---")
        
        # Select only the numerical columns intended for correlation analysis
        numerical_df = self.combined_metadata_df[self.numerical_cols_for_analysis].copy()
        numerical_df = numerical_df.dropna() # Drop rows with any NaN in these columns for correlation

        if numerical_df.empty or len(numerical_df) < 2:
            print("Not enough valid data for correlation analysis after dropping NaNs.")
            return

        correlation_matrix = numerical_df.corr(method='pearson')

        plt.figure(figsize=(12, 10))
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
        plt.title(f'Pearson Correlation Matrix of Numerical Features\n(Source: {self.metadata_dir})')
        plt.tight_layout()
        plt.show()

        print("\nCorrelation Matrix:")
        display(correlation_matrix)

    def perform_statistical_significance_tests(self, alpha=0.05):
        if self.combined_metadata_df is None or self.combined_metadata_df.empty:
            print("No data loaded to perform statistical significance tests.")
            return
        if 'Broad_Genre' not in self.combined_metadata_df.columns:
            print("Error: 'Broad_Genre' column missing. Cannot perform Jazz vs Non-Jazz tests.")
            return

        print(f"\n--- Performing Statistical Significance Tests (Jazz vs. Non-Jazz) for data from: {self.metadata_dir} ---")
        print(f"Significance level (alpha): {alpha}")

        results = {}

        for col in self.numerical_cols_for_analysis: # Use the defined list
            if col not in self.combined_metadata_df.columns:
                print(f"Column '{col}' not found, skipping test.")
                continue

            # Prepare data for the two groups, dropping NaNs for the specific column
            jazz_data = self.combined_metadata_df[self.combined_metadata_df['Broad_Genre'] == 'Jazz'][col].dropna()
            non_jazz_data = self.combined_metadata_df[self.combined_metadata_df['Broad_Genre'] == 'Non-Jazz'][col].dropna()

            if len(jazz_data) < 2 or len(non_jazz_data) < 2: # Need at least 2 samples per group for tests
                print(f"Not enough data for column '{col}' in one or both groups after dropping NaNs. Skipping.")
                results[col] = {"test_used": "Skipped", "statistic": np.nan, "p_value": np.nan, "significant": "N/A", "jazz_normality_p": np.nan, "nonjazz_normality_p": np.nan}
                continue

            # 1. Test for Normality (Shapiro-Wilk test)
            #    H0: The data is normally distributed.
            #    If p-value <= alpha, reject H0 (data is not normal).
            shapiro_jazz_stat, shapiro_jazz_p = stats.shapiro(jazz_data) if len(jazz_data) >=3 else (np.nan, np.nan) # Shapiro needs at least 3 samples
            shapiro_nonjazz_stat, shapiro_nonjazz_p = stats.shapiro(non_jazz_data) if len(non_jazz_data) >=3 else (np.nan, np.nan)

            test_used = ""
            statistic = np.nan
            p_value = np.nan

            # 2. Choose and perform the appropriate test
            # Consider data normal if p > alpha (for normality test)
            # And if sample size is large enough for shapiro (>=3)
            jazz_is_normal = (shapiro_jazz_p > alpha if pd.notna(shapiro_jazz_p) else True) # Assume normal if test couldn't run
            nonjazz_is_normal = (shapiro_nonjazz_p > alpha if pd.notna(shapiro_nonjazz_p) else True)


            if jazz_is_normal and nonjazz_is_normal:
                # If both are normal, check for equal variances (Levene's test)
                # H0: Variances are equal.
                # If p-value <= alpha, reject H0 (variances are not equal).
                levene_stat, levene_p = stats.levene(jazz_data, non_jazz_data)
                equal_var = levene_p > alpha
                
                # Perform t-test
                ttest_stat, ttest_p = stats.ttest_ind(jazz_data, non_jazz_data, equal_var=equal_var, nan_policy='omit')
                test_used = f"T-test (equal_var={equal_var})"
                statistic = ttest_stat
                p_value = ttest_p
            else:
                # If at least one is not normal, use Mann-Whitney U test
                # H0: The distributions of both samples are equal.
                # If p-value <= alpha, reject H0 (distributions are different).
                try:
                    mwu_stat, mwu_p = stats.mannwhitneyu(jazz_data, non_jazz_data, alternative='two-sided', nan_policy='omit')
                    test_used = "Mann-Whitney U"
                    statistic = mwu_stat
                    p_value = mwu_p
                except ValueError as e: # Can happen if all values are identical in one group
                    print(f"Could not perform Mann-Whitney U for {col}: {e}")
                    test_used = "Mann-Whitney U (Error)"


            significant = p_value <= alpha if pd.notna(p_value) else "N/A"
            
            results[col] = {
                "test_used": test_used,
                "statistic": statistic,
                "p_value": p_value,
                "significant_at_alpha_" + str(alpha): significant,
                "jazz_mean": jazz_data.mean(),
                "jazz_median": jazz_data.median(),
                "jazz_std": jazz_data.std(),
                "nonjazz_mean": non_jazz_data.mean(),
                "nonjazz_median": non_jazz_data.median(),
                "nonjazz_std": non_jazz_data.std(),
                "jazz_shapiro_p": shapiro_jazz_p,
                "nonjazz_shapiro_p": shapiro_nonjazz_p
            }
            print(f"\nResults for '{col}':")
            print(f"  Jazz: N={len(jazz_data)}, Mean={jazz_data.mean():.2f}, Median={jazz_data.median():.2f}, Shapiro_p={shapiro_jazz_p:.3f}")
            print(f"  Non-Jazz: N={len(non_jazz_data)}, Mean={non_jazz_data.mean():.2f}, Median={non_jazz_data.median():.2f}, Shapiro_p={shapiro_nonjazz_p:.3f}")
            print(f"  Test Used: {test_used}")
            print(f"  Statistic: {statistic:.3f}, P-value: {p_value:.4f}")
            print(f"  Significant difference at alpha={alpha}: {significant}")

        self.statistical_test_results_df = pd.DataFrame.from_dict(results, orient='index')
        print("\nSummary of Statistical Significance Tests:")
        display(self.statistical_test_results_df)
        return self.statistical_test_results_df


    def generate_all_plots_and_analysis(self): # Renamed for clarity
        if self.combined_metadata_df is None or self.combined_metadata_df.empty:
            print(f"Cannot generate plots or analysis. No metadata loaded for analyzer instance based on {self.metadata_dir}")
            return

        print(f"\n===== Generating All Plots and Analysis for: {self.metadata_dir} =====")
        self.plot_adl_genre_distribution()
        self.plot_piast_yt_tags_distribution()
        self.plot_piast_at_agreement_distribution()
        self.plot_numerical_feature_distributions()
        
        # New analysis calls
        self.perform_correlation_analysis()
        self.perform_statistical_significance_tests()
        
        print(f"===== Finished All Plots and Analysis for: {self.metadata_dir} =====")
```

```python
import pandas as pd
import os
import shutil
import numpy as np
import json

# --- Configuration ---
ORIGINAL_PREPROCESSED_FILES_DIR = r"C:\Users\User\Desktop\college\fyp\datasets\PreprocessedData\validFiles"
ORIGINAL_METADATA_DIR = r"C:\Users\User\Desktop\college\fyp\datasets\PreprocessedData\metadata"

BASE_CLEANED_OUTPUT_DIR = r"C:\Users\User\Desktop\college\fyp\datasets\PreprocessedData_Cleaned" 
CLEANED_VALID_FILES_DIR = os.path.join(BASE_CLEANED_OUTPUT_DIR, "validFiles")
CLEANED_METADATA_DIR = os.path.join(BASE_CLEANED_OUTPUT_DIR, "metadata")

# BPM Threshold
MAX_BPM_THRESHOLD = 300

# <<< MODIFIED: Global Strict Length Thresholds >>>
GLOBAL_MIN_LENGTH_S = 30  # 30 seconds
GLOBAL_MAX_LENGTH_S = 420 # 7 minutes (7 * 60)

# Length Percentile Thresholds (applied *after* global strict length filter)
LENGTH_LOWER_PERCENTILE = 5
LENGTH_UPPER_PERCENTILE = 95
# MIN_ABSOLUTE_LENGTH_S will effectively be GLOBAL_MIN_LENGTH_S after the global filter,
# but we can keep it for the logic within percentile calculation if needed,
# though max(GLOBAL_MIN_LENGTH_S, percentile_lower_bound) will handle it.
MIN_ABSOLUTE_LENGTH_S_FOR_PERCENTILE = GLOBAL_MIN_LENGTH_S # Redundant but keeps logic similar


# Avg Notes/Min Percentile Thresholds
AVG_NOTES_MIN_LOWER_PERCENTILE = 5
AVG_NOTES_MIN_UPPER_PERCENTILE = 95
MIN_ABSOLUTE_AVG_NOTES_MIN = 10
MAX_ABSOLUTE_AVG_NOTES_MIN = 2000

# PIAST related paths
PIAST_AT_TAG_LIST_PATH = r"C:\Users\User\Desktop\college\fyp\datasets\OriginalData\piast\Hayeonbang_PIAST_dataset\piast_at\tag_list.json"
PIAST_AT_CSV_PATH = os.path.join(r"C:\Users\User\Desktop\college\fyp\datasets\OriginalData", "piast", "Hayeonbang_PIAST_dataset", "piast_at", "at_text.csv")
PIAST_YT_JSON_PATH = os.path.join(r"C:\Users\User\Desktop\college\fyp\datasets\OriginalData", "piast", "Hayeonbang_PIAST_dataset", "piast_yt", "yt_text.json")

METADATA_FILENAMES = [
    "adl_jazz_metadata.csv", "adl_non_jazz_metadata.csv",
    "piast_at_jazz_metadata.csv", "piast_at_non_jazz_metadata.csv",
    "piast_yt_jazz_metadata.csv", "piast_yt_non_jazz_metadata.csv",
    "pijama_jazz_metadata.csv"
]

# --- Helper Functions ---
def load_metadata_df(filepath):
    if os.path.exists(filepath):
        try:
            return pd.read_csv(filepath)
        except Exception as e:
            print(f"Error loading metadata file {filepath}: {e}")
    return None

def ensure_dir(directory_path):
    os.makedirs(directory_path, exist_ok=True)

# --- Main Script ---
def filter_files_by_all_criteria_strict_length():
    print("--- Starting Full Anomaly Filtering (BPM, GLOBAL Length, Percentile Length, Avg Notes/Min) ---")
    ensure_dir(CLEANED_VALID_FILES_DIR)
    ensure_dir(CLEANED_METADATA_DIR)

    # Clean output directories from previous run
    if os.path.exists(BASE_CLEANED_OUTPUT_DIR):
        print(f"Removing existing cleaned output directory: {BASE_CLEANED_OUTPUT_DIR}")
        shutil.rmtree(BASE_CLEANED_OUTPUT_DIR)
    ensure_dir(CLEANED_VALID_FILES_DIR)
    ensure_dir(CLEANED_METADATA_DIR)


    all_original_metadata_dfs = {}
    filtered_dfs_for_saving = {}
    valid_file_titles_by_csv = {}

    # 1. Load all original metadata
    print("\n--- Loading Original Metadata ---")
    combined_data_list = []
    for meta_filename in METADATA_FILENAMES:
        original_meta_path = os.path.join(ORIGINAL_METADATA_DIR, meta_filename)
        df = load_metadata_df(original_meta_path)
        if df is None:
            print(f"Skipping {meta_filename} as it could not be loaded.")
            continue

        if "adl" in meta_filename: df['Dataset_Source_Type'] = 'ADL'
        elif "piast_at" in meta_filename: df['Dataset_Source_Type'] = 'PiAST-AT'
        elif "piast_yt" in meta_filename: df['Dataset_Source_Type'] = 'PiAST-YT'
        elif "pijama" in meta_filename: df['Dataset_Source_Type'] = 'PiJAMA'
        else: df['Dataset_Source_Type'] = 'Unknown'

        all_original_metadata_dfs[meta_filename] = df
        combined_data_list.append(df)

    if not combined_data_list:
        print("No metadata loaded. Exiting.")
        return

    full_combined_df = pd.concat(combined_data_list, ignore_index=True)
    print(f"Total files loaded before any filtering: {len(full_combined_df)}")

    # 2. Ensure required columns are numeric
    numeric_cols = ['BPM', 'Length (s)', 'Avg Notes/Min']
    for col in numeric_cols:
        if col in full_combined_df.columns:
            full_combined_df[col] = pd.to_numeric(full_combined_df[col], errors='coerce')
        else:
            print(f"Warning: Column '{col}' not found in combined DataFrame. Filtering for it will be skipped.")


    # 3. Apply BPM filter globally
    print("\n--- Applying BPM Filter ---")
    if 'BPM' in full_combined_df.columns:
        bpm_filtered_df = full_combined_df[
            (full_combined_df['BPM'].notna()) &
            (full_combined_df['BPM'] <= MAX_BPM_THRESHOLD) &
            (full_combined_df['BPM'] > 0)
        ].copy()
        print(f"Files remaining after BPM filter (BPM <= {MAX_BPM_THRESHOLD}): {len(bpm_filtered_df)}")
    else:
        print("BPM column not found. Skipping BPM filter.")
        bpm_filtered_df = full_combined_df.copy()

    # 4. <<< NEW: Apply GLOBAL Strict Length Filter >>>
    print("\n--- Applying GLOBAL Strict Length Filter ---")
    if 'Length (s)' in bpm_filtered_df.columns:
        globally_length_filtered_df = bpm_filtered_df[
            (bpm_filtered_df['Length (s)'].notna()) &
            (bpm_filtered_df['Length (s)'] >= GLOBAL_MIN_LENGTH_S) &
            (bpm_filtered_df['Length (s)'] <= GLOBAL_MAX_LENGTH_S)
        ].copy()
        print(f"Files remaining after GLOBAL strict length filter ({GLOBAL_MIN_LENGTH_S}s - {GLOBAL_MAX_LENGTH_S}s): {len(globally_length_filtered_df)}")
    else:
        print("'Length (s)' column not found. Skipping GLOBAL strict length filter.")
        globally_length_filtered_df = bpm_filtered_df.copy()


    # 5. Apply Percentile-Based Length filter (per-dataset-source percentiles on the already globally filtered data)
    print("\n--- Applying Percentile-Based Length Filter (on globally filtered data) ---")
    length_percentile_filtered_dfs_list = []
    dataset_source_types_len = globally_length_filtered_df['Dataset_Source_Type'].unique()

    for source_type in dataset_source_types_len:
        source_df = globally_length_filtered_df[globally_length_filtered_df['Dataset_Source_Type'] == source_type].copy()
        if source_df.empty or 'Length (s)' not in source_df.columns or source_df['Length (s)'].isnull().all():
            print(f"No valid length data for source type '{source_type}' for percentile filter. Skipping.")
            if not source_df.empty: length_percentile_filtered_dfs_list.append(source_df)
            continue

        length_series = source_df['Length (s)'].dropna() # Already filtered by global min/max
        if len(length_series) < 2:
            print(f"Not enough length data points for '{source_type}' to calculate percentiles. Using existing global bounds for this source.")
            # The data already adheres to GLOBAL_MIN_LENGTH_S and GLOBAL_MAX_LENGTH_S
            # No further percentile filtering possible, so we just keep what passed the global filter for this source type
            length_percentile_filtered_dfs_list.append(source_df)
            continue
        
        lower_b_percentile = np.percentile(length_series, LENGTH_LOWER_PERCENTILE)
        upper_b_percentile = np.percentile(length_series, LENGTH_UPPER_PERCENTILE)
        
        # The effective lower bound is the stricter of the global min and this source's lower percentile
        # Since data is already >= GLOBAL_MIN_LENGTH_S, this max operation is mostly for consistency if
        # MIN_ABSOLUTE_LENGTH_S_FOR_PERCENTILE were different, but here it's the same.
        effective_lower_b = max(lower_b_percentile, MIN_ABSOLUTE_LENGTH_S_FOR_PERCENTILE) # MIN_ABSOLUTE_LENGTH_S_FOR_PERCENTILE is GLOBAL_MIN_LENGTH_S

        print(f"For '{source_type}' (Percentile Length): Range [{effective_lower_b:.2f}s - {upper_b_percentile:.2f}s] (applied on globally filtered data)")
        filtered_subset = source_df[
            (source_df['Length (s)'] >= effective_lower_b) & # .notna() already handled by global filter
            (source_df['Length (s)'] <= upper_b_percentile)
        ]
        length_percentile_filtered_dfs_list.append(filtered_subset)

    if not length_percentile_filtered_dfs_list:
        print("No data remaining after percentile-based length filters. Exiting.")
        return
    intermediate_filtered_df_after_length = pd.concat(length_percentile_filtered_dfs_list, ignore_index=True)
    print(f"Files remaining after percentile-based length filter: {len(intermediate_filtered_df_after_length)}")


    # 6. Apply Avg Notes/Min filter (per-dataset-source percentiles)
    print("\n--- Applying Avg Notes/Min Filter ---")
    final_filtered_dfs_list = []
    dataset_source_types_notes = intermediate_filtered_df_after_length['Dataset_Source_Type'].unique()

    for source_type in dataset_source_types_notes:
        source_df_notes = intermediate_filtered_df_after_length[intermediate_filtered_df_after_length['Dataset_Source_Type'] == source_type].copy()
        if source_df_notes.empty or 'Avg Notes/Min' not in source_df_notes.columns or source_df_notes['Avg Notes/Min'].isnull().all():
            print(f"No valid 'Avg Notes/Min' data for source type '{source_type}'. Skipping this filter for it.")
            if not source_df_notes.empty: final_filtered_dfs_list.append(source_df_notes)
            continue

        notes_min_series = source_df_notes['Avg Notes/Min'].dropna()
        if len(notes_min_series) < 2:
            print(f"Not enough 'Avg Notes/Min' data points for '{source_type}' for percentiles. Applying absolute bounds if defined.")
            lower_n_b = MIN_ABSOLUTE_AVG_NOTES_MIN if MIN_ABSOLUTE_AVG_NOTES_MIN is not None else -np.inf
            upper_n_b = MAX_ABSOLUTE_AVG_NOTES_MIN if MAX_ABSOLUTE_AVG_NOTES_MIN is not None else np.inf
        else:
            lower_n_b = np.percentile(notes_min_series, AVG_NOTES_MIN_LOWER_PERCENTILE)
            upper_n_b = np.percentile(notes_min_series, AVG_NOTES_MIN_UPPER_PERCENTILE)

        if MIN_ABSOLUTE_AVG_NOTES_MIN is not None: lower_n_b = max(lower_n_b, MIN_ABSOLUTE_AVG_NOTES_MIN)
        if MAX_ABSOLUTE_AVG_NOTES_MIN is not None: upper_n_b = min(upper_n_b, MAX_ABSOLUTE_AVG_NOTES_MIN)

        print(f"For '{source_type}' (Avg Notes/Min): Range [{lower_n_b:.2f} - {upper_n_b:.2f}]")
        filtered_subset_notes = source_df_notes[
            (source_df_notes['Avg Notes/Min'].notna()) &
            (source_df_notes['Avg Notes/Min'] >= lower_n_b) &
            (source_df_notes['Avg Notes/Min'] <= upper_n_b)
        ]
        final_filtered_dfs_list.append(filtered_subset_notes)

    if not final_filtered_dfs_list:
        print("No data remaining after all filters. Exiting.")
        return

    master_filtered_df = pd.concat(final_filtered_dfs_list, ignore_index=True)
    print(f"Total files remaining after ALL filters: {len(master_filtered_df)}")


    # 7. Populate valid_file_titles_by_csv for copying and saving filtered metadata
    print("\n--- Preparing Filtered Metadata and File Lists ---")
    # (This part of the logic for mapping back to original CSVs remains the same)
    for meta_filename in METADATA_FILENAMES:
        current_source_type = 'Unknown' # Determine source type based on meta_filename
        if "adl" in meta_filename: current_source_type = 'ADL'
        elif "piast_at" in meta_filename: current_source_type = 'PiAST-AT'
        elif "piast_yt" in meta_filename: current_source_type = 'PiAST-YT'
        elif "pijama" in meta_filename: current_source_type = 'PiJAMA'
        
        relevant_titles = set(master_filtered_df[master_filtered_df['Dataset_Source_Type'] == current_source_type]['Title'].astype(str))
        original_df_for_this_csv = all_original_metadata_dfs.get(meta_filename)

        if original_df_for_this_csv is not None:
            df_to_save = original_df_for_this_csv[original_df_for_this_csv['Title'].astype(str).isin(relevant_titles)].copy()
            if 'Dataset_Source_Type' in df_to_save.columns:
                df_to_save = df_to_save.drop(columns=['Dataset_Source_Type'])
            filtered_dfs_for_saving[meta_filename] = df_to_save
            valid_file_titles_by_csv[meta_filename] = set(df_to_save['Title'].astype(str))
            print(f"For {meta_filename}: Original={len(original_df_for_this_csv)}, After all filters={len(df_to_save)}")
        else:
            valid_file_titles_by_csv[meta_filename] = set()

    # 8. Walk through ORIGINAL_PREPROCESSED_FILES_DIR and copy valid files
    # (This file copying logic remains the same)
    print(f"\n--- Copying Valid MIDI Files to {CLEANED_VALID_FILES_DIR} ---")
    files_copied_count = 0
    files_skipped_count = 0
    if not os.path.exists(ORIGINAL_PREPROCESSED_FILES_DIR):
        print(f"ERROR: Original preprocessed files directory not found: {ORIGINAL_PREPROCESSED_FILES_DIR}")
        return
    for root, _, files in os.walk(ORIGINAL_PREPROCESSED_FILES_DIR):
        for filename in files:
            if not filename.lower().endswith((".mid", ".midi")): continue
            original_filepath = os.path.join(root, filename)
            file_title_no_ext = os.path.splitext(filename)[0]
            relative_path_from_base_files_dir = os.path.relpath(root, ORIGINAL_PREPROCESSED_FILES_DIR)
            path_parts = relative_path_from_base_files_dir.split(os.sep)
            source_csv_key = None
            try:
                if path_parts[0].lower() == "adl": source_csv_key = f"adl_{path_parts[1].lower()}_metadata.csv"
                elif path_parts[0].lower() == "piast" and len(path_parts) > 2: source_csv_key = f"{path_parts[1].lower()}_{path_parts[2].lower()}_metadata.csv"
                elif path_parts[0].lower() == "pijama": source_csv_key = "pijama_jazz_metadata.csv"
            except IndexError: print(f"Warning: Path structure issue for {original_filepath}, cannot map to CSV key.")
            is_valid_for_copy = False
            if source_csv_key and source_csv_key in valid_file_titles_by_csv:
                if file_title_no_ext in valid_file_titles_by_csv[source_csv_key]: is_valid_for_copy = True
            elif not source_csv_key: print(f"Warning: No source CSV key determined for {original_filepath}. File will be skipped.")
            if is_valid_for_copy:
                relative_path_for_output = os.path.relpath(original_filepath, ORIGINAL_PREPROCESSED_FILES_DIR)
                new_filepath = os.path.join(CLEANED_VALID_FILES_DIR, relative_path_for_output)
                new_file_dir = os.path.dirname(new_filepath); ensure_dir(new_file_dir)
                try: shutil.copy2(original_filepath, new_filepath); files_copied_count += 1
                except Exception as e: print(f"Error copying {original_filepath} to {new_filepath}: {e}"); files_skipped_count +=1
            else: files_skipped_count += 1
    print(f"Finished copying: {files_copied_count} files copied, {files_skipped_count} files skipped.")

    # 9. Create new (cleaned) metadata files
    # (This logic for saving metadata remains the same)
    print(f"\n--- Creating Cleaned Metadata Files in {CLEANED_METADATA_DIR} ---")
    for meta_filename, cleaned_df_to_save in filtered_dfs_for_saving.items():
        original_df_len = len(all_original_metadata_dfs.get(meta_filename, pd.DataFrame()))
        cleaned_meta_path = os.path.join(CLEANED_METADATA_DIR, meta_filename)
        try:
            cleaned_df_to_save.to_csv(cleaned_meta_path, index=False, encoding='utf-8')
            print(f"Saved cleaned metadata: {meta_filename} with {len(cleaned_df_to_save)} rows (originally {original_df_len}).")
        except Exception as e: print(f"Error saving cleaned metadata for {meta_filename}: {e}")

if __name__ == "__main__":
    filter_files_by_all_criteria_strict_length() # Renamed function
    print("\n--- Full Anomaly Filtering (with strict length) and New Dataset Creation Complete ---")
```

```python
import os
import shutil
import json
import numpy as np
import pypianoroll
import pretty_midi
from tqdm import tqdm

class AdvancedMidiConverter:
    """
    Converts MIDI files into a rich, multi-channel matrix representation
    that captures onsets, sustains, velocities, and micro-timing offsets.
    """
    def __init__(self, resolution=24, pitch_range=(21, 109)):
        """
        Initializes the advanced converter.

        Args:
            resolution (int): The number of time steps (ticks) per quarter note.
                              Higher values capture finer rhythmic detail. 24 is standard.
            pitch_range (tuple): The MIDI pitch range to process. (21, 109) is an 88-key piano.
        """
        self.resolution = resolution
        self.pitch_range = pitch_range
        self.lowest_pitch, self.highest_pitch = pitch_range[0], pitch_range[1] - 1
        self.num_pitches = self.highest_pitch - self.lowest_pitch + 1
        
        # Define the list of General MIDI program numbers for piano-family instruments (0-7)
        self.PIANO_PROGRAMS = list(range(8))

        print("Advanced MIDI Converter Initialized:")
        print(f"  - Resolution: {self.resolution} ticks per quarter note")
        print(f"  - Pitch Range: {self.pitch_range} ({self.num_pitches} pitches)")

    def midi_to_multichannel_representation(self, midi_filepath):
        """
        The core conversion function. Reads a MIDI, merges piano tracks, and
        extracts the four key matrices: onset, sustain, velocity, and offset.
        """
        try:
            pm = pretty_midi.PrettyMIDI(midi_filepath)
            
            # --- 1. Identify, convert, and collect all piano notes ---
            all_piano_notes = []
            for instrument in pm.instruments:
                # Check if the instrument is in the piano family
                if instrument.program in self.PIANO_PROGRAMS:
                    # Convert Bright Acoustic Piano (1) to Acoustic Grand Piano (0) if needed
                    if instrument.program == 1:
                        instrument.program = 0
                    all_piano_notes.extend(instrument.notes)
            
            if not all_piano_notes:
                return None # Skip files with no piano tracks

            # Sort all collected notes by their start time
            all_piano_notes.sort(key=lambda x: x.start)

            # --- 2. Determine total length and initialize matrices ---
            song_end_time = max(note.end for note in all_piano_notes)
            
            # Calculate ticks per second (sampling frequency for our matrices)
            # This uses the first tempo change, assuming it's constant.
            ticks_per_beat = self.resolution
            tempo_bpm = pm.get_tempo_changes()[1][0] if len(pm.get_tempo_changes()[1]) > 0 else 120.0
            ticks_per_second = (ticks_per_beat * tempo_bpm) / 60.0
            
            total_timesteps = int(np.ceil(song_end_time * ticks_per_second))
            
            # Initialize our four output matrices
            onset_matrix = np.zeros((total_timesteps, self.num_pitches), dtype=np.uint8)
            sustain_matrix = np.zeros((total_timesteps, self.num_pitches), dtype=np.uint8)
            velocity_matrix = np.zeros((total_timesteps, self.num_pitches), dtype=np.float32)
            offset_matrix = np.zeros((total_timesteps, self.num_pitches), dtype=np.float32)

            # --- 3. Populate the matrices ---
            for note in all_piano_notes:
                # Ensure note is within our desired pitch range
                if self.lowest_pitch <= note.pitch <= self.highest_pitch:
                    pitch_idx = note.pitch - self.lowest_pitch
                    
                    # Calculate start and end timesteps
                    onset_step = int(round(note.start * ticks_per_second))
                    offset_step = int(round(note.end * ticks_per_second))

                    # Ensure indices are within bounds
                    if onset_step < total_timesteps:
                        # Onset and Velocity
                        onset_matrix[onset_step, pitch_idx] = 1
                        velocity_matrix[onset_step, pitch_idx] = note.velocity / 127.0
                        
                        # Sustain (Frame)
                        sustain_end = min(offset_step, total_timesteps)
                        sustain_matrix[onset_step:sustain_end, pitch_idx] = 1

                        # Micro-timing Offset
                        quantized_time = onset_step / ticks_per_second
                        time_diff_seconds = note.start - quantized_time
                        timestep_duration = 1.0 / ticks_per_second
                        # Normalize the offset to be between [-0.5, 0.5] of a timestep
                        normalized_offset = time_diff_seconds / timestep_duration
                        offset_matrix[onset_step, pitch_idx] = np.clip(normalized_offset, -0.5, 0.5)

            return {
                "onset": onset_matrix,
                "sustain": sustain_matrix,
                "velocity": velocity_matrix,
                "offset": offset_matrix
            }

        except Exception as e:
            # print(f"Error processing {os.path.basename(midi_filepath)}: {e}")
            return None

    def process_directory(self, input_dir, base_output_dir, label_map, current_label, dataset_group_name):
        """Walks a directory, converts all MIDI files, and saves them."""
        print(f"Processing directory: {input_dir} for group: {dataset_group_name}")
        
        for root, _, files in os.walk(input_dir):
            for filename in tqdm(files, desc=f"Converting {dataset_group_name}"):
                if not filename.lower().endswith((".mid", ".midi")):
                    continue
                
                midi_filepath = os.path.join(root, filename)
                
                # Convert the MIDI file to the multi-channel representation
                multi_channel_data = self.midi_to_multichannel_representation(midi_filepath)

                if multi_channel_data is not None:
                    # Create the output path
                    relative_path = os.path.relpath(root, input_dir)
                    output_sub_dir = os.path.join(base_output_dir, dataset_group_name, relative_path)
                    os.makedirs(output_sub_dir, exist_ok=True)
                    
                    output_npz_filename = os.path.splitext(filename)[0] + ".npz"
                    output_npz_path = os.path.join(output_sub_dir, output_npz_filename)

                    # Save all four matrices into a single compressed .npz file
                    np.savez_compressed(
                        output_npz_path,
                        onset=multi_channel_data["onset"],
                        sustain=multi_channel_data["sustain"],
                        velocity=multi_channel_data["velocity"],
                        offset=multi_channel_data["offset"]
                    )
                    
                    # Update the label map
                    relative_output_path = os.path.relpath(output_npz_path, base_output_dir)
                    label_map[relative_output_path.replace('\\', '/')] = current_label
        
        return label_map

# ==============================================================================
# Main Execution Block
# ==============================================================================
if __name__ == "__main__":
    BASE_INPUT_MIDI_DIR = r"C:\Users\User\Desktop\college\fyp\datasets\PreprocessedData_Cleaned\validFiles"
    # Create a new output directory for this advanced representation
    BASE_OUTPUT_MATRIX_DIR = r"C:\Users\User\Desktop\college\fyp\datasets\PianoRollMatrices"

    # --- Configuration ---
    RESOLUTION = 24  # Standard ticks per quarter note
    PITCH_LOW = 21   # A0
    PITCH_HIGH = 108 # C8
    PITCH_RANGE = (PITCH_LOW, PITCH_HIGH + 1)
    
    # Instantiate the new converter
    converter = AdvancedMidiConverter(
        resolution=RESOLUTION,
        pitch_range=PITCH_RANGE
    )

    datasets_to_process = [
        {"input_path": os.path.join(BASE_INPUT_MIDI_DIR, "adl", "jazz"), "label": 1, "group_name": os.path.join("adl", "jazz")},
        {"input_path": os.path.join(BASE_INPUT_MIDI_DIR, "adl", "non_jazz"), "label": 0, "group_name": os.path.join("adl", "non_jazz")},
        {"input_path": os.path.join(BASE_INPUT_MIDI_DIR, "piast", "piast_at", "jazz"), "label": 1, "group_name": os.path.join("piast", "piast_at", "jazz")},
        {"input_path": os.path.join(BASE_INPUT_MIDI_DIR, "piast", "piast_at", "non_jazz"), "label": 0, "group_name": os.path.join("piast", "piast_at", "non_jazz")},
        {"input_path": os.path.join(BASE_INPUT_MIDI_DIR, "piast", "piast_yt", "jazz"), "label": 1, "group_name": os.path.join("piast", "piast_yt", "jazz")},
        {"input_path": os.path.join(BASE_INPUT_MIDI_DIR, "piast", "piast_yt", "non_jazz"), "label": 0, "group_name": os.path.join("piast", "piast_yt", "non_jazz")},
        {"input_path": os.path.join(BASE_INPUT_MIDI_DIR, "piJAMA"), "label": 1, "group_name": os.path.join("piJAMA", "jazz")}
    ]

    # Clean output directory
    if os.path.exists(BASE_OUTPUT_MATRIX_DIR):
        print(f"Warning: Output directory {BASE_OUTPUT_MATRIX_DIR} exists. Removing it for a fresh run.")
        shutil.rmtree(BASE_OUTPUT_MATRIX_DIR)
    os.makedirs(BASE_OUTPUT_MATRIX_DIR, exist_ok=True)

    master_label_map = {}
    for dataset_info in datasets_to_process:
        converter.process_directory(
            input_dir=dataset_info["input_path"],
            base_output_dir=BASE_OUTPUT_MATRIX_DIR,
            label_map=master_label_map,
            current_label=dataset_info["label"],
            dataset_group_name=dataset_info["group_name"]
        )

    # Save the master label map
    label_map_path = os.path.join(BASE_OUTPUT_MATRIX_DIR, "master_label_map.json")
    with open(label_map_path, 'w') as f:
        json.dump(master_label_map, f, indent=4)
    print(f"\n--- Conversion Complete. Master label map saved to: {label_map_path} ---")

    # --- Example: How to Load and Inspect the New Data ---
    print("\n--- Example: Loading a new multi-channel .npz file ---")
    if master_label_map:
        example_relative_path = list(master_label_map.keys())[0]
        example_full_path = os.path.join(BASE_OUTPUT_MATRIX_DIR, example_relative_path)
        
        print(f"Loading example file: {example_full_path}")
        with np.load(example_full_path) as data:
            print("Matrices found in file:", list(data.keys()))
            for key in data.keys():
                print(f"  - Shape of '{key}' matrix: {data[key].shape}")
```

```python
import os
import numpy as np
import pretty_midi
import scipy.io.wavfile
from tqdm import tqdm

# ==============================================================================
# Cell: Convert a single NPZ to MIDI and WAV using a Python-native approach
# ==============================================================================

# --- 1. CONFIGURATION: UPDATE THESE PATHS ---

# The multi-channel NPZ file you want to convert.
NPZ_FILE_PATH = r"C:\Users\User\Desktop\college\fyp\datasets\PianoRollMatrices\adl\non_jazz\The Nutcracker Op71-atto 1 - Marche.npz"

# [IMPORTANT] The full path to your SoundFont file.
SOUNDFONT_PATH = r"C:\Users\User\Desktop\college\fyp\other\soundfont\GeneralUser GS v1.471.sf2"

# Where to save the output files.
OUTPUT_DIR = r"C:\Users\User\Desktop\college\fyp\converted_samples"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# These parameters MUST match the ones used during data creation.
RESOLUTION = 24         # Ticks per quarter note
PITCH_LOW = 21          # MIDI pitch of the lowest key (A0)
BPM = 120.0             # The default BPM to interpret the timesteps

# --- END OF CONFIGURATION ---


def convert_npz_to_wav(npz_path, soundfont_path, output_dir):
    """
    Loads a multi-channel NPZ, reconstructs a PrettyMIDI object,
    saves it as a .mid file, and synthesizes it to a .wav file using pretty_midi.
    """
    if not os.path.exists(npz_path):
        print(f"ERROR: Input NPZ file not found at '{npz_path}'")
        return
        
    if not os.path.exists(soundfont_path):
        print(f"ERROR: SoundFont file not found at '{soundfont_path}'")
        print("Please download a .sf2 file (like GeneralUser GS) and update the SOUNDFONT_PATH.")
        return

    # --- 1. Load the Multi-Channel Data ---
    print(f"Loading data from {os.path.basename(npz_path)}...")
    with np.load(npz_path) as data:
        onset_matrix = data['onset']
        sustain_matrix = data['sustain']
        velocity_matrix = data['velocity']
        offset_matrix = data['offset']
    
    print("Onset Matrix shape:", onset_matrix.shape)
    print("Sustain Matrix shape:", sustain_matrix.shape)
    print("Velocity Matrix shape:", velocity_matrix.shape)
    print("Offset Matrix shape:", offset_matrix.shape)


    print("Onset Matrix unique values:", np.unique(onset_matrix))
    print("Sustain Matrix unique values:", np.unique(sustain_matrix))
    print("Velocity Matrix unique values:", np.unique(velocity_matrix))
    print("Offset Matrix unique values:", np.unique(offset_matrix))
    
    print("Data loaded successfully. Reconstructing MIDI...")

    # --- 2. Reconstruct the MIDI object from matrices ---
    pm = pretty_midi.PrettyMIDI(resolution=RESOLUTION, initial_tempo=BPM)
    instrument = pretty_midi.Instrument(program=0, name='Acoustic Grand Piano')
    
    tick_duration = 60.0 / (BPM * RESOLUTION)
    onset_coords = np.where(onset_matrix == 1)
    
    for t_step, pitch_idx in tqdm(zip(*onset_coords), total=len(onset_coords[0]), desc="Reconstructing notes"):
        quantized_start_time = t_step * tick_duration
        time_offset = offset_matrix[t_step, pitch_idx] * tick_duration
        start_time = quantized_start_time + time_offset
        
        end_step = t_step
        while end_step < sustain_matrix.shape[0] and sustain_matrix[end_step, pitch_idx] == 1:
            end_step += 1
        end_time = end_step * tick_duration
        
        if end_time <= start_time:
            end_time = start_time + tick_duration
            
        pitch = pitch_idx + PITCH_LOW
        velocity = int(velocity_matrix[t_step, pitch_idx] * 126) + 1
        
        note = pretty_midi.Note(
            velocity=max(1, min(127, velocity)), # Ensure velocity is in valid [1, 127] range
            pitch=pitch,
            start=start_time,
            end=end_time
        )
        instrument.notes.append(note)

    pm.instruments.append(instrument)
    
    # --- 3. Save to MIDI file (good for debugging and reuse) ---
    filename_base = os.path.splitext(os.path.basename(npz_path))[0]
    output_midi_path = os.path.join(output_dir, f"{filename_base}_reconstructed.mid")
    pm.write(output_midi_path)
    print(f"MIDI file saved to: {output_midi_path}")

    # --- 4. Synthesize MIDI to WAV using pretty_midi's internal synthesizer ---
    print("Synthesizing MIDI to WAV audio data...")
    sample_rate = 44100  # Standard CD-quality audio
    
    try:
        # This calls FluidSynth on the backend to generate the audio waveform as a NumPy array
        audio_data = pm.fluidsynth(fs=sample_rate, sf2_path=soundfont_path)
    except Exception as e:
        print("\n--- SYNTHESIS ERROR ---")
        print(f"An error occurred during synthesis: {e}")
        print("Please ensure FluidSynth is installed and accessible in your system's PATH.")
        return

    # --- 5. Save the audio data as a WAV file ---
    output_wav_path = os.path.join(output_dir, f"{filename_base}_reconstructed.wav")
    
    # The output of .fluidsynth is a float array from -1 to 1.
    # We need to convert it to 16-bit integers for standard WAV format.
    scaled_audio = np.int16(audio_data / np.max(np.abs(audio_data)) * 32767)
    
    scipy.io.wavfile.write(output_wav_path, sample_rate, scaled_audio)
    print(f"Successfully synthesized and saved WAV file to: {output_wav_path}")


# --- Main Execution Block ---
if __name__ == "__main__":
    convert_npz_to_wav(NPZ_FILE_PATH, SOUNDFONT_PATH, OUTPUT_DIR)
```

```python
import os
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

# ==============================================================================
# 1. Configuration
# ==============================================================================
BASE_MATRIX_DIR = r"C:\Users\User\Desktop\college\fyp\datasets\PianoRollMatrices"
MASTER_LABEL_MAP_FILE = os.path.join(BASE_MATRIX_DIR, "master_label_map.json")

# ==============================================================================
# 2. Main Analysis Function
# ==============================================================================
def analyze_musicality_of_jazz_files(base_dir, label_map_file, segment_length=128, window_size=8):
    """
    [MODIFIED] Analyzes musicality metrics on a SEGMENT level, ignoring any
    segments that are completely empty, exactly like the DataLoader.

    Args:
        base_dir (str): The base directory where piano roll matrices are stored.
        label_map_file (str): Path to the JSON file mapping file paths to labels.
        segment_length (int): The length of segments to analyze (e.g., 128).
        window_size (int): The size of the sliding window for note rate analysis.
    """
    print("--- Starting Musicality Analysis on NON-EMPTY SEGMENTS ---")
    print(f"Segment Length: {segment_length} timesteps | Sliding Window: {window_size} timesteps")

    # --- Step 1: Load the label map and filter for jazz files ---
    print(f"Loading label map from: {label_map_file}")
    with open(label_map_file, 'r') as f:
        label_map = {k.replace('\\', '/'): v for k, v in json.load(f).items()}

    jazz_files_relative = [path for path, label in label_map.items() if label == 1]
    if not jazz_files_relative:
        print("Error: No jazz files (label=1) found in the label map.")
        return

    print(f"Found {len(jazz_files_relative)} jazz files to scan for segments.")
    
    # --- Step 2: Iterate through files, segment them, and calculate metrics on non-empty segments ---
    all_densities = []
    all_max_notes_in_window = []
    all_min_notes_in_window = []
    all_avg_notes_in_window = [] # [NEW] List to store the average rate per segment
    
    for relative_path in tqdm(jazz_files_relative, desc="Scanning files for non-empty segments"):
        full_path = os.path.join(base_dir, relative_path)
        
        try:
            with np.load(full_path) as data:
                onset_matrix = data['onset']
                num_timesteps = onset_matrix.shape[0]

            for start_step in range(0, num_timesteps - segment_length + 1, segment_length):
                onset_segment = onset_matrix[start_step : start_step + segment_length, :]

                if np.sum(onset_segment) == 0:
                    continue
                
                # Metric 1: Note Density
                all_densities.append(np.mean(onset_segment))

                # Metric 2, 3, 4: Max, Min, and Avg Notes in a Sliding Window
                onsets_per_timestep = np.sum(onset_segment, axis=1)
                kernel = np.ones(window_size)
                notes_in_window = np.convolve(onsets_per_timestep, kernel, mode='same')
                
                all_max_notes_in_window.append(np.max(notes_in_window))
                all_min_notes_in_window.append(np.min(notes_in_window))
                all_avg_notes_in_window.append(np.mean(notes_in_window)) # [NEW] Calculate and store the average

        except Exception:
            pass

    if not all_densities:
        print("\nError: Could not find any non-empty segments to analyze. Please check paths and data.")
        return

    print(f"\nAnalysis complete. Found and analyzed {len(all_densities)} non-empty segments.")

    # --- Step 3: Calculate and print statistics ---
    all_densities = np.array(all_densities)
    all_max_notes_in_window = np.array(all_max_notes_in_window)
    all_min_notes_in_window = np.array(all_min_notes_in_window)
    all_avg_notes_in_window = np.array(all_avg_notes_in_window) # [NEW] Convert to numpy array

    # Density Stats
    print("\n" + "="*50)
    print("--- Note Density Analysis Results ---")
    mean_density = np.mean(all_densities)
    print(f"Mean Density:                {mean_density:.4f}")
    print(f"Median Density:              {np.median(all_densities):.4f}")
    print(f"\nSuggests a TARGET_DENSITY_END is ~{mean_density:.4f}.")
    
    # [NEW] Average Note Rate Stats
    print("\n" + "="*50)
    print(f"--- [NEW] Average Note Rate (per window) Analysis ---")
    mean_avg_rate = np.mean(all_avg_notes_in_window)
    median_avg_rate = np.median(all_avg_notes_in_window)
    print(f"Mean of Average Rates:       {mean_avg_rate:.2f}")
    print(f"Median of Average Rates:     {median_avg_rate:.2f}")
    print("\nThis metric describes the 'typical busyness' of a segment.")

    # Max Notes in Window Stats
    print("\n" + "="*50)
    print(f"--- Max Notes in Window Analysis ---")
    median_max_notes = np.median(all_max_notes_in_window)
    p90_max_notes = np.percentile(all_max_notes_in_window, 90)
    print(f"Median of Maximums:          {median_max_notes:.2f}")
    print(f"90th Percentile:             {p90_max_notes:.2f}")
    print(f"\nSuggests a MAX_NOTES_IN_WINDOW_END around {median_max_notes:.0f} or {p90_max_notes:.0f}.")

    # Minimum Notes in Window Stats
    print("\n" + "="*50)
    print(f"--- Min Notes in Window Analysis ---")
    median_min_notes = np.median(all_min_notes_in_window)
    p25_min_notes = np.percentile(all_min_notes_in_window, 25)
    print(f"Median of Minimums:          {median_min_notes:.2f}")
    print(f"25th Percentile:             {p25_min_notes:.2f}")
    print(f"\nSuggests a MIN_NOTES_IN_WINDOW_THRESHOLD for your penalty could be {p25_min_notes:.0f} or {median_min_notes:.0f}.")
    print("="*50)

    # --- Step 4: Plot the distributions ---
    
    plt.figure(figsize=(16, 12)) # [NEW] Adjusted figure size for a 2x2 grid
    plt.suptitle('Musicality Analysis of Non-Empty Jazz Segments', fontsize=20, y=1.03)

    # Plot 1: Density
    plt.subplot(2, 2, 1) # [NEW] Changed to 2x2 grid
    sns.histplot(all_densities, bins=50, kde=True, color='skyblue')
    plt.axvline(np.mean(all_densities), color='r', linestyle='--', label=f'Mean: {np.mean(all_densities):.4f}')
    plt.title('Distribution of Note Density')
    plt.xlabel('Note Onset Density')
    plt.ylabel('Number of Segments')
    plt.legend()

    # Plot 2: Average Note Rate [NEW PLOT]
    plt.subplot(2, 2, 2) # [NEW] Changed to 2x2 grid
    sns.histplot(all_avg_notes_in_window, bins=50, kde=True, color='purple')
    plt.axvline(np.mean(all_avg_notes_in_window), color='r', linestyle='--', label=f'Mean: {np.mean(all_avg_notes_in_window):.2f}')
    plt.title('Distribution of Average Notes per Window')
    plt.xlabel(f'Average Notes per {window_size}-step Window')
    plt.ylabel('Number of Segments')
    plt.legend()

    # Plot 3: Max Notes in Window
    plt.subplot(2, 2, 3) # [NEW] Changed to 2x2 grid
    sns.histplot(all_max_notes_in_window, discrete=True, color='salmon')
    plt.axvline(np.mean(all_max_notes_in_window), color='r', linestyle='--', label=f'Mean: {np.mean(all_max_notes_in_window):.2f}')
    plt.title(f'Distribution of Max Notes per Window')
    plt.xlabel(f'Maximum Notes in any {window_size}-step Window')
    plt.legend()
    
    # Plot 4: Minimum Notes in Window
    plt.subplot(2, 2, 4) # [NEW] Changed to 2x2 grid
    sns.histplot(all_min_notes_in_window, discrete=True, color='lightgreen')
    plt.axvline(np.mean(all_min_notes_in_window), color='r', linestyle='--', label=f'Mean: {np.mean(all_min_notes_in_window):.2f}')
    plt.title(f'Distribution of Min Notes per Window')
    plt.xlabel(f'Minimum Notes in any {window_size}-step Window')
    plt.xticks(np.arange(0, int(np.max(all_min_notes_in_window)) + 2, 1))
    plt.xlim(-1, 15)
    plt.legend()

    plt.tight_layout(rect=[0, 0, 1, 0.98]) # Adjust layout to make space for suptitle
    plt.show()
    plt.close()


# ==============================================================================
# 3. Execution Block
# ==============================================================================
if __name__ == "__main__":
    analyze_musicality_of_jazz_files(
        base_dir=BASE_MATRIX_DIR,
        label_map_file=MASTER_LABEL_MAP_FILE,
        segment_length=128, # Should match your DataLoader's segment_length
        window_size=64       # Should match your loss function's window_size
    )
```
